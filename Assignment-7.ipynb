```python
import numpy as np
import pandas as pd
roll_no=int(input("Enter your roll number"))
np.random.seed(roll_no)
sales_data=np.random.randint(1000,5001,size=(12,4))
months=['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories=['Electronics', 'Cloth' , 'Home&Kitchen', 'Sports']
sales_df=pd.DataFrame(sales_data, columns=categories, index=months)
print(sales_df)
```

```python
print("FIRST FIVE ROWS:")
print(sales_df.head())
print("\nSUMMARY STATISTICS:")
print(sales_df.describe())
```

```python
tot_sales_cat=sales_df.sum()
tot_sales_month=sales_df.sum(axis=1)
print("\nCategory-wise total sales:\n",tot_sales_cat)
print("\nMonth-wise total sales:\n",tot_sales_month)
```

```python
sales_growth=sales_df.pct_change()*100
avg_growth=sales_growth.mean()
print("\nAverage Sales Growth:\n", avg_growth)
sales_df['Total Sales']=sales_df.sum(axis=1)
sales_df['Growth Rate']=sales_df['Total Sales'].pct_change() * 100
print("before discount, total sales month-wise:\n",tot_sales_month)
```

```python
if roll_no%2==0:
    sales_df['Electronics']*=0.90
else:
    sales_df['Cloth']*=0.85

#tot_sales_month=sales_df.sum(axis=1)
sales_df['Total Sales']=sales_df.sum(axis=1)
print("\n after applying discount, total sales month-wise:\n",tot_sales_month)

print("\nUpdated Sales Dataframe:")
print(sales_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
for category in categories:
    plt.plot(sales_df.index, sales_df[category], marker='o', label=category)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.grid()
plt.show()

# Box plots for sales distribution
plt.figure(figsize=(8, 6))
sns.boxplot(data=sales_df[categories])
plt.title("Sales Distribution by Category")
plt.ylabel("Sales")
plt.show()
```

```python
import numpy as np
array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_array = np.abs(array)
print("Element-wise absolute value:")
print(abs_array)
 
percentiles_flt= np.percentile(array, [25, 50, 75])
percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)

print("\nPercentiles for flattened array (25th, 50th, 75th):")
print(percentiles_flt)
print("\nPercentiles for each column:")
print(percentiles_col)
print("\nPercentiles for each row:")
print(percentiles_rows)


mean_flt = np.mean(array)
median_flt = np.median(array)
std_flt = np.std(array)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

mean_rows = np.mean(array, axis=1)
median_rows = np.median(array, axis=1)
std_rows = np.std(array, axis=1)

print("\nMean, Median, and Standard Deviation for flattened array:")
print(f"Mean: {mean_flt}")
print(f"Median: {median_flt}")
print(f"Std Dev: {std_flt}")

print("\nMean, Median, and Standard Deviation for each column:")
print(f"Mean: {mean_col}")
print(f"Median: {median_col}")
print(f"Std Dev: {std_col}")

print("\nMean, Median, and Standard Deviation for each row:")
print(f"Mean: {mean_rows}")
print(f"Median: {median_rows}")
print(f"Std Dev: {std_rows}")
```

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
print("Floor values:", floor_values)
ceiling_values = np.ceil(a)
print("Ceiling values:", ceiling_values)
truncated_values = np.trunc(a)
print("Truncated values:", truncated_values)
rounded_values = np.round(a)
print("Rounded values:", rounded_values)
```

```python
def swap_elements(list, index1, index2):
    temp = list[index1]  
    list[index1] = list[index2]  
    list[index2] = temp  
    return list

list = [10, 20, 30, 40, 50]
index1, index2 = 1, 3
swapped_list = swap_elements(list, index1, index2)
print("List after swapping:", swapped_list)
```

```python
def swap(st,idx1,idx2):
    lst = list(st)
    lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
    swapped_set = set(lst)
    return swapped_set
st = {10,20,40,30,50}
print("Original Set:", st)
swapped_set = swap(st,2,3)
print("Set after swapping:", swapped_set)
```
