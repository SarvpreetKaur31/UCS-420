```python
import numpy as np 
x=np.linspace(0,2*np.pi,100)
y=np.linspace(0,2*np.pi,100)
z=np.sin(x)
a=np.cos(y)
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x,z)
plt.plot(y,a)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, z, label="Sine Wave") 
plt.plot(y, a, label="Cos Wave") 
plt.title("Sine Wave and Cos Wave")
plt.legend()
plt.show()
```

```python
#QUESTION-2
arr= np.array([10, 52, 62, 16, 16, 54, 453])
sorted=np.sort(arr)
print("Sorted Array is:",sorted)
indices=np.argsort(arr)
print("Indices of elements in sorted array:",indices)
min4=sorted[:4]
print("4 smallest elements:",min4)
max5=sorted[-5:]
print("5 largest elements:",max5)

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
int_ele=array[array==array.astype(int)]
print("integer elements in array:",int_ele)
float_ele=array[array!=array.astype(int)]
print("float elements in array:",float_ele)
```

```python
# QUESTION-3
first='S'
last='K'
X=ord(first)+ord(last)
sales=np.array([X,X+50,X+100,X+150,X+200])
print("Ascii Sum:",first,"+",last,"=",X)
print("Sales Dataset:",sales)

tax_rate=((X % 5) + 5) / 100
print("tax rate:",tax_rate*100,"%")
tax_sales=sales*(1+tax_rate)
print("sales after tax:",tax_sales)

dsct_rate=np.where(sales<X+100,0.05,0.1)
dsct_sales=sales*(1-dsct_rate)
print("discount rate:",dsct_rate)
print("final sales after discount:",dsct_sales)

sales_matrix=np.tile(sales,(3,1))
weekly_increase= np.array([1.00,1.02,1.04]).reshape(3,1)
adjusted_sales=sales_matrix*weekly_increase
print("original sales matrix:\n",sales_matrix)
print("sales matrix after weekly increase:\n",adjusted_sales)
```

```python
#QUESTION-4
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
y1=x**2
y2=np.sin(x)
y3=np.exp(x)
y4=np.log(np.abs(x)+1)
plt.figure(figsize=(12, 8))
#y1
plt.subplot(2,2,1)
plt.plot(x,y1,'r',label=r'$Y = x^2$')
plt.title("Quadratic Function: $Y = x^2$'")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
#y2
plt.subplot(2,2,2)
plt.plot(x,y2,'g',label=r'$Y = sin(x)$')
plt.title("Sine Function: $Y = sin(x)$")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
#y3
plt.subplot(2,2,3)
plt.plot(x,y3,'b',label=r'$Y = e^x$')
plt.title("Exponential Function: $Y = e^x$")
plt.xlabel('X')
plt.ylabel('Y')
#plt.yscale('log')
plt.grid(True)
plt.legend()
#y4
plt.subplot(2,2,4)
plt.plot(x,y4,'m',label=r'$Y = log(|x| + 1)$')
plt.title("Logarithmic Function: $Y = log(|x| + 1)$")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
