```python
import numpy as np
import matplotlib.pyplot as plt
M=float(input("Enter the value of m"))
x=np.linspace(-10,10,100)
y1=M*x*x
y2=M*np.sin(x)
plt.plot(x,y1,'b--', label='y=M*x^2')
plt.plot(x,y2,'g', label='y=M*Sin(x)')
plt.legend()
plt.grid()
plt.title("Mathematical Functions")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data={'Subject':['English', 'Maths', 'Science', 'Punjabi','History'], 'Score': [90, 100, 100, 80, 96]}
df=pd.DataFrame(data)
plt.figure(figsize=(8,5))
ax = sns.barplot(x='Subject', y='Score', data=df, hue='Subject', palette="pastel", legend=False)
for p in ax.patches:
    ax.annotate(f"{p.get_height()}", (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

plt.title("Scores in Subjects", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12, fontweight='bold')
plt.ylabel("Scores", fontsize=12, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
import numpy as np
import matplotlib.pyplot as plt
roll_no=int(input("Enter your roll number as seed"))
np.random.seed(roll_no)
data=np.random.randn(50)
#print(data)
fig,ax=plt.subplots(2, 2, figsize=(10,8))
ax[0, 0].plot(np.cumsum(data), marker='o', linestyle='-', color='b')
ax[0, 0].set_title("Cumulative Sum Line Plot", fontweight='bold', fontsize=16)
ax[0, 0].set_xlabel("Index")
ax[0, 0].set_ylabel("Cumulative Sum")
ax[0, 0].grid(True)

ax[0, 1].scatter(range(50), data, color='r', alpha=0.7)
ax[0, 1].set_title("Scatter Plot with Random Noise", fontweight='bold', fontsize=16)
ax[0, 1].set_xlabel("Index")
ax[0, 1].set_ylabel("Value")
ax[0, 1].grid(True)
ax[1, 0].axis("off")
ax[1, 1].axis("off")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\SARVPREET KAUR\Downloads\company_sales_data.csv")
sns.set_theme(style="whitegrid")
#Lineplot
plt.figure(figsize=(10, 5))
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker='o', color='b')
plt.title("Total Profit Over All Months", fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
#Multilineplot
plt.figure(figsize=(12, 6))
for column in df.columns[1:-1]:
    sns.lineplot(x=df["month_number"], y=df[column], label=column, marker='o')
plt.title("Product Sales Over Months", fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Products")
plt.grid(True)
plt.show()
#Barchart
plt.figure(figsize=(12, 6))
df.drop(columns=["month_number"]).sum().plot(kind="bar", colormap="coolwarm")
plt.title("Total Sales and Profit for All Features", fontweight='bold')
plt.xlabel("Attributes")
plt.ylabel("Total Sales/Profit")
plt.grid(axis='y', linestyle="--")
plt.show()
```
